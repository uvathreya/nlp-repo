```python
from collections import Counter
import nltk
```

```python
text = " Online learning platforms are booming day by day. providing multiple ways to learn from. one such example is udemy."
lower_case = text.lower()
tokens = nltk.tokenize.word_tokenize(lower_case)
tags = nltk.pos_tag(tokens)
print("tags-> ", tags)
```

```python
counts = Counter( tag for word, tag in tags)
print(counts)
```

```python
fd = nltk.FreqDist(tokens)
fd.plot()
```
